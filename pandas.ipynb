```python
#identifying null values variables
isnull = df.isnull()
isnull
```

```python
isnull_count = isnull.sum()
print("Total number of null values:", isnull_count)
```

```python
#total nulls n dataset
total_isnull_count = isnull.sum().sum()
print("Total nulls:", total_isnull_count)
```

```python
#replace single quote with blank and make the date formatted
df['Date'] = df['Date'].replace("'", "", regex = True)
df
```

```python
#handle null values, return dataframe with no empty cells
#we can use df.dropna(), it removes empty cells but will not change the data frame
#if we want dataframe to be changed, simply use inplace= True
new_df = df.dropna()
new_df
```

```python
isnull_count_newdf = isnull.sum()
print("Total number of null values:", isnull_count_newdf)
```

```python
isnull_count_newdf = new_df.isnull().sum().sum()
print("Total nulls:", isnull_count_newdf)
```

```python
#another way of handling the null values is by replacing the null values
df.fillna(130,inplace=True)
df
```
